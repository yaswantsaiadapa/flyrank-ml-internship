## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

My chosen lane, **Content Refresh Opportunity Scoring**, is best framed as a **ranking/scoring** machine learning task.

The goal is not simply to determine whether a webpage should be refreshed, but to prioritize webpages based on their expected need for a content refresh. The model assigns each webpage a priority score, and these scores are used to rank pages from highest to lowest priority.

This ranking helps SEO managers decide which pages to refresh first when editorial resources are limited. A ranking/scoring approach is more appropriate than simple classification because it produces an ordered list of refresh opportunities rather than only a yes/no decision.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

The goal of this project is to produce a **content refresh priority score** for each webpage. Since the starter dataset does not contain a direct label indicating the true refresh priority, the project uses a **proxy** based on observed historical performance signals.

The proxy is informed by measurable SEO and engagement metrics, such as impressions, click-through rate, engagement, and content freshness, which help identify pages that are likely to benefit from a refresh.

This project does not rely on an arbitrary manually defined rule for refresh priority. Instead, it uses observed data to support ranking webpages for editorial review and decision-making.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

The primary success metric for this project is **Precision@K**, where **K** represents the number of webpages the SEO team can realistically refresh in a given period.

Precision@K measures how many of the top-ranked webpages recommended by the model are truly high-priority refresh opportunities. Since editorial teams have limited time and resources, correctly identifying the most valuable pages is more important than ranking every page perfectly.

A good model would achieve a **high Precision@K**, meaning that most of the webpages in the top recommendations are genuinely worthwhile candidates for content refresh. This ensures the model provides useful, decision-support recommendations for SEO managers.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A fixed rule cannot reliably identify the best content refresh opportunities because refresh priority depends on many interacting SEO and engagement signals rather than a single condition.

For example, rules such as "refresh pages older than one year" or "refresh pages with high impressions" ignore important factors like click-through rate, average search position, engagement, and content performance. Two webpages with similar impressions may require different decisions because their overall performance differs.

Machine learning can analyze these multiple features simultaneously and learn complex patterns from historical data. Instead of relying on numerous manually defined if-else rules, the model assigns a refresh priority score based on the combined influence of many signals, producing more accurate and scalable recommendations for SEO managers.
